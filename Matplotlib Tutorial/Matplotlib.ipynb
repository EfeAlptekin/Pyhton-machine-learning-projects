```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Plot y versus x as lines and/or markers.
plt.plot([1,2,3,4], [1,4,9,16])
```

```python
x = [1,2,3,4]
y = [1,4,9,16]


plt.plot(x,y)
plt.show()
```

```python
plt.title("Title")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show
```

```python
#if we dont prefer the ticks,pyhton puts the values automatically
#But we can also prefer the values

plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
#Let we put legend
plt.plot(x,y, label='x^2',color='green')
plt.xticks=([1,2,3,4,5])
plt.yticks=([1,4,9,16,25])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label='x^2',color='green',linewidth = 2,linestyle = '--',marker='.')
plt.xticks=([1,2,3,4,5])
plt.yticks=([1,4,9,16,25])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label='x^2',color='green',linewidth = 2,linestyle = '--',marker='.')
plt.xticks=([1,2,3,4,5])
plt.yticks=([1,4,9,16,25])
plt.xlabel('x axis')
plt.ylabel('y axis')

x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color='red',linewidth='2',marker='.',label='2*x')


plt.legend()
plt.show()
```

```python
#Barchart
x = ['Ankara', 'İstanbul', 'İzmir']
y = [120, 178, 87]

plt.bar(x,y)
plt.show()
```

```python
x = ['Ankara', 'İstanbul', 'İzmir']
y = [120, 178, 87]

cubuklar = plt.bar(x,y)
cubuklar[1].set_hatch('/')
cubuklar[0].set_hatch('.')
plt.show()
```

```python
gas = pd.read_csv('petrolfiyatlari.csv')

gas
```

```python
plt.title('Oil Prices')
plt.plot(gas['Year'],gas['USA'],'b-',label='USA')
plt.xlabel('Year')
plt.ylabel('Dolar')

plt.legend()
plt.show()
```

```python
plt.title('Oil Prices')

plt.plot(gas.Year, gas.USA, 'b-', label='USA')
plt.plot(gas.Year, gas['Canada'], 'r-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas['France'], 'y.-', label='France')
plt.plot(gas['Year'], gas.Germany, color='orange', label='Germany')

plt.xlabel('Yıl')
plt.ylabel('Dolar')

plt.legend()

plt.show()
```

```python
#Congiguration the size of table
plt.figure(figsize=(14,5))
plt.title('Oil Prices')

plt.plot(gas.Year, gas.USA, 'b-', label='USA')
plt.plot(gas.Year, gas['Canada'], 'r-', label="Canada")
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas['France'], 'y.-', label='France')


plt.xlabel('Yıl')
plt.ylabel('Dolar')

plt.legend()
plt.savefig('ikincigrafigim.png', dpi=300)
plt.show()
```
